### 1. Post-smoothing 시행 

```python
# 2. 각 컬럼 데이터프레임에서 -1-1-1-1 1 -1-1-1- 일 경우, 1 을 -1으로 만들어 주는 함수 destory 

def remove(talk_column):
    """
    각 컬럼 데이터프레임에서 -1,-1,-1, 1, -1, -1, -1 일 경우, 1을 -1로 만들어 주는 함수
    
    """
    
    dest = []
    for i, v in enumerate(talk_column.values):
        if i == 0:
            pass
        elif(talk_column.values[i-1] + talk_column.values[i] == 0 and talk_column.values[i] + talk_column.values[i+1] == 0):
            if talk_column.values[i] == 1.0:
                dest.append(i)
    print(dest) #어떤 인덱스에서 1 이 나오는가? 
    for i in dest:
        talk_column.iloc[i, :] = talk_column.iloc[i, :].replace(1,-1)
    
    return talk_column
```

```python
# # 3. 각 컬럼을 destory 한 다음에 talk_column 을 원본 데이터의 talk1, talk2, talk3 대신 넣어준다. 

# new = pd.concat([destroy(T1), destroy(T2), destroy(T3)], axis = 1)
# dff.drop(['talk1', 'talk2', 'talk3'], axis = 1)
# new_df = pd.concat([dff, new], axis = 1)
# new_df[:3]
```

```python
def post_smth(original_df):
    """
    p1, p2, p3 컬럼 별로 remove를 완료하여 새로운 데이터프레임 생성. turn-taking 디택션 함수의 input이 됨 
    """
    test_aa = original_df['talk1']
    test_bb = original_df['talk2']
    test_cc = original_df['talk3']

    t1 = pd.DataFrame(test_aa).reset_index(drop = True)
    t2 = pd.DataFrame(test_bb).reset_index(drop = True)
    t3 = pd.DataFrame(test_cc).reset_index(drop = True)

    new = pd.concat([remove(t1), remove(t2), remove(t3)], axis = 1)
    original_df = original_df.drop(['talk1', 'talk2', 'talk3'], axis = 1)
    new_df = pd.concat([original_df, new], axis = 1)
    
    return new_df
```

## 3. person 별, 연속된 침묵 구간 및 음성 구간 길이를 측정.  

```python
## <example> #####
# p1 --> 함수변환 #
# -1       -1    #    
# -1       -2    #
# +1        1    #
# +1        2    #
# -1       -1    #
# -1       -2    #
##################

def detect_talk_break_length(talk):  #p1 의 talk1 열만 들어간다. 
    '''
    input:  pandas data frame column of one person's talk.
    output: python list of one person's dialog length which includes break_length(negative) and talk_length(positive)
    '''
    total = 0
    dialog_len = []
    for i in range(len(talk)): #0 - 50662
        prev_total = total
        total += talk[i]
        #if i >= 5445 and i <=5600:
            #print("id", i, "prev:", prev_total, "total", total)
        if(total < 0 and total > prev_total):
            dialog_len.append(prev_total)
            total = 1
        
        elif(total > 0 and total < prev_total):
            dialog_len.append(prev_total)
            total = -1
        if(i==len(talk)-1):
            dialog_len.append(total)# 마지막 누계는 무조건 append
    
    return dialog_len
```

## 4. 턴테이킹 함수 

```python
def dup(guest_dup):
    a = pd.DataFrame(guest_dup)
    b = (a.iloc[:,0] == 1.0).sum()
    return b
```

```python
# 수정 완료 

only_talk = new_df.iloc[:, 8:11].copy()
def get_start_end_times(dialog_len_host, dialog_len_guest):
    """
    
    - dialog_len_host 에서 dialog_len_guest 로의 턴테이킹만 측정
    - return 되는 값? 
        * host의 음성구간 시작점-끝점
        * guest의 음성구간 시작점-끝점
        * guest의 음성구간 길이 --> short-response 가려내기 위함
        
    """
    start = 0
    start_end = []
    for i in range(len(dialog_len_host)-1):  
        start += abs(dialog_len_host[i])
        if dialog_len_host[i] < 0: # a: host 음성구간 시작 인덱스, b : host 음성구간 끝 인덱스
            a = start                           
            b = (start + abs(dialog_len_host[i + 1]) - 1)
            start_end.append([a,b])
    

    start2 = 0  
    start_end2 = [] 
    guest_talk_times = []
    for i in range(len(dialog_len_guest)-1):
        start2 += abs(dialog_len_guest[i])
        if dialog_len_guest[i] < 0: # c: guest 음성구간 시작 인덱스, d : guest 음성구간 끝 인덱스 
            c = start2
            d = (start2 + abs(dialog_len_guest[i + 1]) - 1)
            start_end2.append([c,d])
            
        else:
            guest_talk_times.append(dialog_len_guest[i])
    
    
    return start_end, start_end2, guest_talk_times
```

```python
# host_end1 = start_end
# 
# guest_number = 0,1,2 (p1, p2, p3)

def count_turn_takes(host_number, guest_number, host_start_end, guest_start_end, guest_talk_times1, gap=300): #숏 리스펀스 
    """
    
    host_number = host가 되는 person의 인덱스 => p1 p2 p3 == 0 1 2 
    guset_number = guest가 되는 person의 인덱스 => p1 p2 p3 == 0 1 2
    host_start_end = get_start_end_times 함수 결과의 start_end (host 음성 구간의 시작 인덱스, 끝 인덱스)
    guest_start_end = get_start_end_times 함수 결과의 start_end2 (guest 음성 구간의 시작 인덱스, 끝 인덱스)
    guest_talk_times1 = get_start_end_times 함수 결과의 guest_talk_times
    gap = short response 길이 상한선
    
    """
    short_responses = 0
    total_responses = 0
    change = []
    for i in range(len(host_start_end)):
        k = 0

        while k<=len(guest_start_end)-1 and guest_start_end[k][0] < host_start_end[i][1]:   #k +=1 결과가 len(guest_start_end)보다는 작거나 같아야 한다. 
            k += 1   # i를 돌면서 k를 업데이트, 각 i 별로 +1 
            

        #print(i, k,host_start_end[i], guest_start_end[k])
        if k<=len(guest_start_end)-1 and guest_start_end[k][0] -host_start_end[i][1] <= 300 and  guest_start_end[k][0] -host_start_end[i][1] > 0: # 디폴트: 300 이내에 말 해야함
            
            ## host 가 말하는 중에는 guest가 침묵하고 있어야함. 다만, total 500 정도 말한 것은, 호응/미러링일 수 있으므로, 이 경우는 턴테이킹
            if dup(only_talk.iloc[int(host_start_end[i][0]): int(host_start_end[i][1])+1, guest_number]) <= 500:
                
                ## guest 가 말하는 중에는 host가 침묵하고 있어야함. 다만, total 500 정도 말한 것은, 미러링일 수 있으므로, 턴테이킹으로 분류
                if dup(only_talk.iloc[int(guest_start_end[k][0]): int(guest_start_end[k][1])+1, host_number]) <= 500:
                
                    if guest_talk_times1[k] <= gap:
                        short_responses += 1
                        change.append(k)
                    else:
                        total_responses += 1
                        change.append(k)
                        
    return total_responses, short_responses, change
```

## 5. Silence detection : 누가 먼저 침묵을 깼는가? 횟수 측정

```python
# Person1, Person2, Person3 gap > 300:
#     silence
```

```python
def table(whois): 
    """creating a final table for silence breaker
    """
    z = list(itertools.chain(*whois))
    df3 = pd.DataFrame.from_dict(Counter(z), orient='index').reset_index()
    df3.columns = ['person', 'interruption']
    df3.sort_values(by='person', ascending=True, inplace = True)
    df_final = df3.copy()
    df_final.loc[:, 'person'].replace(0, 'person1', inplace = True)
    df_final.loc[:, 'person'].replace(1, 'person2', inplace = True)
    df_final.loc[:, 'person'].replace(2, 'person3', inplace = True)
    return df_final
```

```python
def combi_suspect(DataFrame):
    """
        - creating a two-columned dataframe
        - 1st column : a row corresponding to the index of p1&p2&p3, being silent
            - each row of the 1st column value is -1, meaning silence
            - 1st column consists of original index of the Scenario file
        - 2nd column : 
            - continuously adding -1 as the line goes by
            - stop accumulating -1 when the index jumps more than 2 steps
            - it means there is a silence breaker         
    """ 
    
    
    silent_times = []
    for i in range(len(DataFrame)):
        if (DataFrame['talk1'][i]+DataFrame['talk2'][i] + DataFrame['talk3'][i]) == -3:
            silent_times.append(i)   

    pd_silence = pd.DataFrame(silent_times.copy())
    pd_silence.columns = ['idx']
    pd_silence

# silence 옆에 추가할, -1 accumulation 한 column 생성 
    acc = []
    total = 0
    silence = 0

    for idx, value in enumerate(pd_silence.iloc[:,0]):
        total += silence
        if(value == 0):
            silence = -1
            acc.append(silence)
        elif(pd_silence.iloc[idx, 0] - pd_silence.iloc[idx-1, 0] >= 2):
            silence = -1
            acc.append(silence)
        elif(pd_silence.iloc[idx, 0] - pd_silence.iloc[idx-1, 0] < 2):
            silence += -1
            acc.append(silence)
        
    acc_silence = pd.DataFrame(acc.copy())
    acc_silence.columns = ['acc_idx']
    comb_silence = pd.concat([pd_silence, acc_silence], axis = 1)    

    return comb_silence
```

## 6. Mirroring 횟수 측정 : ex. p1이 말하는 중에 p2이 짧게 반응하는 횟수

```python
def start_to_end(p1_dialog_len):
    """
    p# 이 말한 구간들에 대하여, [start_talk_index ,  end_talk_index] 을 하나의 pair로 구하여 list 에 append한다.  
    """
    start_time = 0
    end_time = 0
    start_end = []
    for i in range(len(p1_dialog_len)-1):  #host 의 start point를 구한다. 
        start_time += abs(p1_dialog_len[i])
        
        if p1_dialog_len[i] < 0:
            a = start_time + 1                            # a = start_time, b = end_time --> [a, b]를 묶은 것을 사전에 정의한 list에 넣어라  
            b = start_time + abs(p1_dialog_len[i + 1])
            start_end.append([a,b])
    return start_end
        
```

```python
def short_pair(P1, P2, how_long):
    """
    P1 이 말하는 중에, P2가 미러링하는 횟수를 구함
    P2가 how_long 이하로 말해야지 미러링. how_long을 넘어가면 미러링이 아님.
    """
    short_response = 0
    for i, value in enumerate(P1):
        for k, value2 in enumerate(P2):
            if(abs(value2[0] - value2[1])<=how_long):
                if(value[0] <= value2[0] and value[1] >= value2[1]):
                    short_response += 1 

    return short_response
```

```python
def mirroring(dialog1, dialog2, dialog3, how_long = 500):
    """
    
    - p1, p2, p3 중에서 가능한 모든 pair에 대해 mirroring 횟수를 구함
    - 결과 순서: 
        p1 --> p2
        p1 --> p3
        p2 --> p1
        p2 --> p3
        p3 --> p1
        p3 --> p2
        
    """

    p1 = start_to_end(dialog1)
    p2 = start_to_end(dialog2)
    p3  = start_to_end(dialog3)
    case = []
    items = [p1, p2, p3]

    # 모든 경우의 수 1,2 != 2,1
    for i in list(permutations(items, 2)):
        case.append(i)
    
    result = []
    for idx, v in enumerate(case):
        a = short_pair(case[idx][0], case[idx][1], how_long)
        result.append(a)
        
        a = pd.DataFrame(result)
        b = pd.DataFrame(['p1 -> p2', 'p1 -> p3', 'p2 -> p1', 'p2 -> p3', 'p3 -> p1', 'p3 -> p2'])
        
        result_mirroring = pd.concat([b, a], axis = 1)
        result_mirroring.columns = ['order', '# of mirroring']
        
    return result_mirroring
    
```
